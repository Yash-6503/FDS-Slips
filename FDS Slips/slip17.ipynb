```python
#slip17_q1
# Write a Python program to draw scatter plots to compare two features of the iris dataset
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("../Datasets/iris.csv")

petal_length=data['petal.length']
petal_width=data['petal.width']

plt.scatter(petal_length, petal_width, color="red")
plt.title("Comparison of two features of iris")
plt.show()
```
