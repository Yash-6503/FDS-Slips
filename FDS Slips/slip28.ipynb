```python
#slip28_q1
#. To create a dataframe containing columns name, age and percentage. Add 10 rows to the dataframe. View the dataframe. 
import pandas as pd

# Create the DataFrame
data = {'name': ['John', 'Emma', 'Michael', 'Sophia', 'Daniel', 'Olivia', 'Matthew', 'Emily', 'Jacob', 'Ava'],
        'age': [25, 30, 28, 22, 35, 27, 32, 29, 26, 31],
        'percentage': [85.5, 92.3, 77.8, 88.9, 91.2, 79.6, 84.1, 88.7, 90.4, 82.9]}

df = pd.DataFrame(data)

# View the DataFrame
print(df)
```

```python
#slip28_q2
# To print the shape, number of rows-columns, data types, feature names and the description of the data.

# Print the shape
print("Shape of the DataFrame:", df.shape)

# Print the number of rows and columns
print("Number of rows:", len(df))
print("Number of columns:", len(df.columns))

# Print the data types
print("Data types:\n", df.dtypes)

# Print the feature names
print("Feature names:", df.columns.tolist())

# Print the description of the data
print("Description of the data:\n", df.describe())
```
