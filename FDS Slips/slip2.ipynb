```python
#slip2_q1  Write a Python program for Handling Missing Value. Replace missing value of salary, age column with mean of that column.(Use Data.csv file).
import pandas as pd

# Load the data from the CSV file
data = pd.read_csv('../Datasets/Data.csv')

# Calculate the mean of the 'salary' column
salary_mean = data['salary'].mean()

# Replace the missing values in the 'salary' column with the mean
data['salary'].fillna(salary_mean, inplace=True)

# Calculate the mean of the 'age' column
age_mean = data['age'].mean()

# Replace the missing values in the 'age' column with the mean
data['age'].fillna(age_mean, inplace=True)

# Save the updated data to a new CSV file
data.to_csv('Updated_Data.csv', index=False)

#atlast check the csv file whether changes occurs or not
```

```python
#slip2_q2  Write a Python program to generate a line plot of name Vs salary 
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('../Datasets/Data.csv')

# Plot the line graph
plt.plot(data['name'], data['salary'], color='blue', linestyle='-')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Name vs Salary')
plt.show()
```
