```python
#slip16_q1
#Create two lists, one representing subject names and the other representing marks obtained in those subjects. Display the data in a pie chart. 
import matplotlib.pyplot as plt

subject_names=['maths', 'physics', 'biology', 'chemistry', 'english']

marks_obtained=['82','55','75','67','90']

plt.pie(marks_obtained, labels=subject_names, autopct='%1.1f%%')
plt.show()
```

```python
#slip16_q2
#Write a python program to create a data frame for students’ information such as name, graduation percentage and age. Display average age of students, average of graduation percentage.
import pandas as pd
import numpy as np
df=pd.DataFrame({
    'name':['Yash', 'Harsh', 'Vivek', 'Vinay', 'Abhi'],
    'age':[20, 21, 25, 19, 27],
    'per':[90, 80, 85, 60, 75]
})
print(np.average(df['age']))
print(np.average(df['per']))
```
