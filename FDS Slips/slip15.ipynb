```python
#slip15_q1
#Generate a random array of 50 integers and display them using a line chart, scatter plot, histogram and box plot. Apply appropriate color, labels and styling options
import matplotlib.pyplot as plt 
import numpy as np
np.random.seed(42)
data=np.random.randint(1, 100, 50)

#line Chart
plt.figure(figsize=(10,5))
plt.plot(data, color="blue", linestyle="-")
plt.title("Line chart")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True)
plt.show()

#Scatter Plot
plt.figure(figsize=(10,5))
plt.scatter(range(len(data)), data, color="red")
plt.title("Scatter Plot")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True)
plt.show()

#Histogram
plt.figure(figsize=(10,5))
plt.hist(data, bins=10, color="blue")
plt.title("Histogram")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()

#box plot
plt.figure(figsize=(10,5))
plt.boxplot(data, vert=False, widths=0.6)
plt.title("Box Plot")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```
