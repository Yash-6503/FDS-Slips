```python
#slip25_q1
# Generate a random array of 50 integers and display them using a line chart, scatter plot, histogram and box plot. Apply appropriate color, labels and styling options. 
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
data=np.random.randint(1, 100, 50)

#line chart
plt.figure(figsize=(10,5))
plt.plot(data, color="red", linestyle='-')
plt.title("Line Chart")
plt.xlabel("index")
plt.ylabel("value")
plt.grid(True)
plt.show()

#Scatter Plot
plt.figure(figsize=(10,5))
plt.scatter(range(len(data)), data, color="green")
plt.title("Scatter Plot")
plt.xlabel("index")
plt.ylabel("value")
plt.grid(True)
plt.show()

#Histogram
plt.figure(figsize=(10,5))
plt.hist(data, bins=10, color='orange')
plt.title("Histogram")
plt.xlabel("index")
plt.ylabel("value")
plt.show()

#Box Plot
plt.figure(figsize=(10,5))
plt.boxplot(data, vert=False, widths=0.6)
plt.title("Box Plot")
plt.xlabel("index")
plt.ylabel("value")
plt.show()
```

```python
#slip25_q2
# Create two lists, one representing subject names and the other representing marks obtained in those subjects. Display the data in a pie chart. 
import matplotlib.pyplot as plt

# Create the lists for subject names and marks obtained
subject_names = ['Math', 'Science', 'English', 'History']
marks_obtained = [90, 85, 95, 80]

# Create a pie chart
plt.pie(marks_obtained, labels=subject_names, autopct='%1.1f%%')

# Add a title to the chart
plt.title('Marks Obtained in Subjects')

# Display the chart
plt.show()
```
