```python
#slip18_q1
# Write a Python program to create box plots to see how each feature i.e. Sepal Length, Sepal Width, Petal Length, Petal Width are distributed across the three species. (Use iris.csv dataset)
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('../Datasets/iris.csv')
new_data=data[['sepal.length','sepal.width','petal.length','petal.width']]
#print(new_data)
#box plot
plt.figure(figsize=(8,4))
new_data.boxplot(vert=False, widths=0.6)
plt.title("Box Plot")
plt.show()
```
