```python
#slip3_q1
#Write a Python program to create box plots to see how each feature i.e. Sepal Length, Sepal Width, Petal Length, Petal Width are distributed across the three species. (Use iris.csv dataset) 

import pandas as pd
import matplotlib.pyplot as plt

# Read the iris dataset into a pandas DataFrame
df = pd.read_csv('../Datasets/iris.csv')

# Create box plots for each feature across the three species
df.boxplot(column=['sepal.length', 'sepal.width', 'petal.length', 'petal.width'], by='variety')


# Set the labels for the x-axis and y-axis
plt.xlabel('Features')
plt.ylabel('Measurements')

# Show the plot
plt.show()
```
