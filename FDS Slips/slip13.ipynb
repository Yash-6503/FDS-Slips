```python
#slip13_q1
#Write a Python program to create a graph to find relationship between the petal length and petal width.(Use iris.csv dataset)
import pandas as pd
import matplotlib.pyplot as plt

# Load the iris dataset
df = pd.read_csv('../Datasets/iris.csv')

# Extract the petal length and petal width columns
petal_length = df['petal.length']
petal_width = df['petal.width']

# Create a scatter plot
plt.scatter(petal_length, petal_width, color="red")

# Set labels and title
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Relationship between Petal Length and Petal Width')

# Show the plot
plt.show()
```

```python
#slip13_q2
#Write a Python program to find the maximum and minimum value of a given flattened array.
import numpy as np
demo=np.array([[0,1],[2,3]])
print("original flattened array is:",demo)
maximum=demo.max()
print("maximum value is:",maximum)
minimum=demo.min()
print("minimum value is:",minimum)
```

```python
#Another way to solve max and min flattended array
# Define the flattened array
flattened_array = [5, 2, 9, 1, 7, 3]

# Find the maximum value
maximum_value = max(flattened_array)

# Find the minimum value
minimum_value = min(flattened_array)

# Print the results
print("Maximum value:", maximum_value)
print("Minimum value:", minimum_value)
```
