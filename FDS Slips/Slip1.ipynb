```python
#slip1_q1 Write a Python program to create a Pie plot to get the frequency of the three species of the Iris data (Use iris.csv) 
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a DataFrame
data_df = pd.read_csv('../Datasets/iris.csv')

# Get the counts of each species
species_counts = data_df['variety'].value_counts()

# Create a pie plot
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')

# Add a title
plt.title('Frequency of Iris Species')

# Display the plot
plt.show()
```
