```python
#slip24_q1
# Import dataset “iris.csv”. Write a Python program to create a Bar plot to get the frequency of the three species of the Iris data
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("../Datasets/iris.csv")
#species frequency
species_frequency=data['variety'].value_counts()
#bar plot
plt.bar(species_frequency.index, species_frequency.values)
plt.title("Bar Plot")
plt.xlabel('index')
plt.ylabel('value')
plt.show()
```

```python
#slip24_q2
#Write a Python program to create a histogram of the three species of the Iris data.
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("../Datasets/iris.csv")
species=data['variety']
#histogram
plt.hist(species, bins=10, color="orange")
plt.title("Histogram")
plt.xlabel("index")
plt.ylabel("value")
plt.show()
```
