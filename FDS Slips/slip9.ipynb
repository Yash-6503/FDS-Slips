```python
#slip9_q1
#Generate a random array of 50 integers and display them using a line chart, scatter plot. Apply appropriate color, labels and styling options
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
data=np.random.randint(1, 100, 50)

#line chart
plt.figure(figsize=(10, 5))
plt.plot(data, color="red", linestyle="-")
plt.title("Line chart")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid()
plt.show()

#Scatter plot
plt.figure(figsize=(8, 4))
plt.scatter(range(len(data)), data, color="green")
plt.title("Scatter Plot")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
#slip9_q2
#Create two lists, one representing subject names and the other representing marks obtained in those subjects. Display the data in a pie chart. 

import matplotlib.pyplot as plt

# Subject names list
subject_names = ['Math', 'Science', 'English', 'History', 'Geography']

# Marks obtained list
marks_obtained = [90, 85, 92, 88, 95]

# Create a pie chart
plt.pie(marks_obtained, labels=subject_names, autopct='%1.1f%%')

# Add a title
plt.title('Marks Obtained in Subjects')

# Display the chart
plt.show()
```
